#### Data Dictionary
The data is about customers of the treadmill product(s) of a retail store called Cardio Good Fitness. It contains the following variables-

* Product - The model no. of the treadmill
* Age -  Age of the customer in no of years
* Gender - Gender of the customer
* Education - Education of the customer in no. of years
* Marital Status - Marital status of the customer
* Usage - Avg. # times the customer wants to use the treadmill every week
* Fitness - Self rated fitness score of the customer (5 - very fit, 1 - very unfit)
* Income - Income of the customer
* Miles- Miles that a customer expects to run

### Conclusion and Recommendations

#### Key insights/observations

* Most users fall between age of 25 to 35 and buy product category - TM195 which is most popular and sold product. Median age of customer is 28. Most user run less than 200 miles and Income varies between 30k to 60k. There is a healthy mix of both male, female buyer. TM498 is lesser popular product compared to TM195.

* People with Income greater than $70000 buy TM798. People with more education make more income and generally buy the TM798 treadmill. Their usage is also greater than 3 times a week which is higher than other products. They also tend to have fitness score between 4-5 rating. Mostly Males bought this product and have fitness score of 5.0 and age between 25 to 35 compared to women. Partnered people tend to run more than Single within this group. 
Females have not bought TM798 product 7 compared to 29 males. 

#### Recommendation for customer profiling:

Fitness programs can be introduced to maintain strong customer retention as the focus of this user group is fitness and they are frequent and regular users. TM798 is the premium pricer product supposedly. Improve product for new opportunities is to analyze female interest. Marital Status is more likely to purchase.
* TM195: 
Target both male and female users with income greater than 30,000 and less than 60,000 within age range of 25-35 years with TM195,TM498 product.
